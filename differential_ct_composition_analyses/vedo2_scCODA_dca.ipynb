```python
#Import all the required packages 
import pandas as pd
import csv
import pickle as pkl
import matplotlib.pyplot as plt
from sccoda.util import cell_composition_data as dat
from sccoda.util import data_visualization as viz
from sccoda.util import comp_ana as mod
import os
```

```python
# Define the parental directory where you save the output results 
par_dir = '/groups/umcg-weersma/tmp01/projects/vedopredict2/biopsies/ongoing/compositional_analysis/scCODA/results/batch1/'
```

```python
#file for the input
file = '/groups/umcg-weersma/tmp01/projects/vedopredict2/biopsies/ongoing/compositional_analysis/scCODA/inputs/batch1/Vedo2_Batch1_inflammation_&_timepoint.csv'
#conditions that you want to test
conditions = ['T0-I_vs_T0-NI', 'T4-I_vs_T4-NI', 'T4-NI_vs_T0-NI', 'T4-I_vs_T0-I'] 
#fdrs per conditions that you want to consider 
fdrs = [0.05, 0.1, 0.3, 0.5]
```

```python
# read the input file
data = pd.read_csv(file, sep=',')
# remove 1st non named column
data = data.iloc[: , 1:]
```

```python
data_subsetted = data
# specify the name of the column used as covariate
covariate = 'state_and_timepoint'


for condition in conditions:
    #per condition, create the name of the new output folder 
    out = condition + '/'
    #output directory: parental directory + the name of the out directory
    output_dir = par_dir + out
    #variable for checking the existance of the output dir 
    isExist = os.path.exists(output_dir)
    #if the output_dir doesn't exist, create the directory
    if not isExist:
        os.mkdir(output_dir)
    #extract from condition name, the treatment name
    treatment = condition.split('_')[0]
    #extract the control name
    control = condition.split('_')[2]
    #transform the pandas dataframe, in object for scCODA
    data_all = dat.from_pandas(data_subsetted, covariate_columns=[covariate])
    #subset the scCODA object with the conditions you need 
    data_sub = data_all[data_all.obs[covariate].isin([treatment , control])]
    #Create the barplot
    viz.boxplots(data_sub, feature_name=covariate)
    plt.tight_layout()
    #save the barplot in the output directory
    plt.savefig(output_dir + condition + '.pdf')
    # perform the compositional analysis, using control as reference 
    model = mod.CompositionalAnalysis(data_sub, formula="C(" + covariate + ", Treatment('" + control + "'))")
    #prepare the results 
    results = model.sample_hmc()
    #iterate over the fdrs that you want to consider 
    for fdr in fdrs:
        #prepare the summary of the results 
        x = results.summary_prepare([fdr])
        x1 = x[1]['log2-fold change']
        lfc = list(x1.values)
        cellnames = []
        for name in x1.index:
            final_name = name[1]
            cellnames.append(final_name)
                
        #path of your final table output 
        path= output_dir + condition + str(fdr)
        #result table part 1
        part1 = x[0]
        #result table part 2
        part2 = x[1]
        #save the two tables in the output directory
        part1.to_csv(output_dir + condition + str(fdr) + 'part_1.tsv', sep= "\t")
        part2.to_csv(output_dir + condition + str(fdr) + 'part_2.tsv', sep= "\t")



```
